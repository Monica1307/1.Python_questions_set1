```python
5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number

```

```python
num = int(input("Enter a number: "))
count = 0
for i in range(1, num + 1):
    if num % i == 0:
        count += 1
if count == 2:
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```
